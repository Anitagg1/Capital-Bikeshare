```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bike_day = pd.read_csv('bike_day.csv')
```

Visualization and first insights

```python
# causal customers rent the most on weekends
sns.catplot(x="weekday", y="casual", hue="year", kind="swarm", data=bike_day)
```

```python
# registered customers rent throughout the working week more than during the weekends, although the difference is not big. 
sns.catplot(x="weekday", y="registered", hue="year", kind="swarm", data=bike_day)
```

```python
# there is been an increase of registered customers from year 2011 to 2012
sns.catplot(x="month", y="registered", hue="year", kind="swarm", data=bike_day)
```

```python
# registered customers are double the casual customers
cols_plot = ['casual', 'registered']
axes = bike_day[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Total reservations')
```

```python
# Nobody rents when the weather is strong rain and snow
sns.catplot(x="weathersit", y="registered", hue="year", kind="swarm", data=bike_day)
```

```python
#FOr 2011 and 2012 there very few rainy days
bike_day['weathersit'].value_counts(ascending=False)
```

```python
bike_day.groupby('weathersit')['year'].value_counts().sort_index(ascending=True)
```

```python
#Registered customers use the service mostly during workdays but very much (in comparison with casual customers) as well during weekend and holidays.

sns.catplot(x="workingday", y="registered", hue="year", kind="swarm", data=bike_day)
```

```python
sns.catplot(x="workingday", y="casual", hue="year", kind="swarm", data=bike_day)
```

Analysis of 2011

```python
# only day bike info for 2011
bikes_2011 = bike_day[bike_day.year.isin(['2011'])]
```

```python
sns.catplot(x="weathersit", y="registered", hue="year", kind="swarm", data=bikes_2011)
```

```python
# only day bike info for 2012
bikes_2012 = bike_day[bike_day.year.isin(['2012'])]
```

```python
sns.catplot(x="weathersit", y="registered", hue="year", kind="swarm", data=bikes_2012)
```

```python
import seaborn as sns

sns.set(rc={'figure.figsize':(11,8)})

sns.set_context("paper", font_scale=1, rc={"lines.linewidth": 2.5})
#sns.set(style="ticks", context="talk")
#plt.style.use("dark_background")
sns.set_style("dark")

#fig, ax = plt.subplots()
ax = sns.pointplot(data = bikes_2012[['month','registered','season']], x = 'month', y = 'registered',
              scale = 0.2, hue = 'season');

plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='16') # for legend title

ax.set(title = 'Season wise daily distribution of bike rentals',ylabel= 'total_count');
```

```python
sns.set(rc={'figure.figsize':(11,8)})

sns.set_context("paper", font_scale=1, rc={"lines.linewidth": 2.5})
#sns.set(style="ticks", context="talk")
#plt.style.use("dark_background")
sns.set_style("dark")

#fig, ax = plt.subplots()
ax = sns.pointplot(data = bikes_2012[['month','registered','weathersit']], x = 'month', y = 'registered',
              scale = 0.2, hue = 'weathersit');

plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='16') # for legend title

ax.set(title = 'Season wise daily distribution of bike rentals',ylabel= 'total_count');
```
